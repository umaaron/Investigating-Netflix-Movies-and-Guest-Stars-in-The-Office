```python
import matplotlib.pyplot as plt
import pandas as pd
data= pd.read_csv("datasets/office_episodes.csv") ## read data
data.info() ## data information 
data.head(5)
```

```python
## color scheme 
cols = []

for ind, row in data.iterrows():
    if row['scaled_ratings'] < 0.25:
        cols.append('red')
    elif row['scaled_ratings'] < 0.50:
        cols.append('orange')
    elif row['scaled_ratings'] < 0.75:
        cols.append('lightgreen')
    else:
        cols.append('darkgreen')
                
cols
```

```python
## sizeing system 

sizes =[]

for ind, row in data.iterrows():
    if row['has_guests'] == False: 
        sizes.append(25)
    else:
        sizes.append(250)
        
sizes       
```

```python
###
fig = plt.figure()
plt.scatter(data.episode_number,data. viewership_mil,c=cols,s=sizes)
plt.xlabel('Episode Number')
plt.ylabel('Viewership (Millions)')
plt.title('Popularity, Quality, and Guest Appearances on the Office')
plt.show()
```

```python
data['colors']=cols

data['sizes']=sizes

data.info()
```

```python
non_guest_data=data[data['has_guests'] ==False]
guest_data= data[data['has_guests']==True]
```

```python
###

fig = plt.figure()
plt.style.use('fivethirtyeight')
plt.rcParams['figure.figsize'] = [20, 17]
plt.scatter(non_guest_data.episode_number,non_guest_data.viewership_mil,
            c=non_guest_data.colors,
            s=non_guest_data.sizes)

plt.scatter(guest_data.episode_number,guest_data.viewership_mil,
            c=guest_data.colors,
            s=guest_data.sizes, 
            marker= "*"
            )



plt.xlabel('Episode Number')
plt.ylabel('Viewership (Millions)')
plt.title('Popularity, Quality, and Guest Appearances on the Office')
plt.show()
```
